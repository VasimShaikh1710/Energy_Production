# Heatmap

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

# Scatter Plot

# Histogram

```python
sns.distplot(df['amb_pressure'], label='amb_pressure',bins = 10)

plt.xlabel('Values of amb_pressure')
plt.ylabel('Frequency')
plt.legend()
plt.title('Histogram of amb_pressure')
plt.show()
```

# Box Plot

```python
df.boxplot(column = 'temperature', vert = False)
```

```python
df.boxplot(column = 'exhaust_vacuum', vert = False)
```

```python
df.boxplot(column = 'amb_pressure', vert = False)
```

```python
import numpy as np
```

```python
aQ1 = np.percentile(df['amb_pressure'],25)
aQ3 = np.percentile(df['amb_pressure'], 75)

aIQR = aQ3 - aQ1

print(f"aQ1 = {aQ1}\naQ3 = {aQ3}\naIQR = {np.round(aIQR)}")
```

```python
LW = aQ1 - (1.5*aIQR)
UW = aQ3 + (1.5*aIQR)

print(f"LW = {LW}\nUW = {UW}")
```

```python
a1 = df[df["amb_pressure"] < LW]
a1
```

```python
a2 = df[df["amb_pressure"] > UW]
a2
```

```python
df.boxplot(column = 'r_humidity', vert = False)
```

```python
rQ1 = np.percentile(df['r_humidity'],25)
rQ3 = np.percentile(df['r_humidity'], 75)

rIQR = rQ3 - rQ1

print(f"rQ1 = {rQ1}\nrQ3 = {rQ3}\nrIQR = {np.round(rIQR)}")
```

```python
LW = Q1 - (1.5*IQR)
UW = Q3 + (1.5*IQR)

print(f"LW = {LW}\nUW = {UW}")
```

```python
r1 = df[df["r_humidity"] < LW]
r1
```

```python
df[df["r_humidity"] > UW]
```

```python
df.boxplot(column = 'energy_production', vert = False)
```

```python
out = pd.concat([a1,a2,r1])
out
```
